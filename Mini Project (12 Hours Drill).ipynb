```python
class Flower:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def __str__(self):
        return f"{self.name} - Price: ${self.price} - In Stock: {self.stock}"

    def sell(self, quantity):
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        else:
            return False

    def order_more(self, quantity):
        self.stock += quantity
        print(f"Ordered {quantity} more {self.name}.")

class Bouquet:
    def __init__(self):
        self.flowers = []

    def add_flower(self, flower, quantity):
        self.flowers.append((flower, quantity))

    def calculate_total_price(self):
        total_price = 0
        for flower, quantity in self.flowers:
            total_price += flower.price * quantity
        return total_price

    def sell(self):
        for flower, quantity in self.flowers:
            if not flower.sell(quantity):
                print(f"Insufficient stock for {flower.name}. Order more before selling.")
                return False
        return True

if __name__ == "__main__":
    rose = Flower("Rose", 2.5, 10)
    tulip = Flower("Tulip", 1.5, 20)
    daisy = Flower("Daisy", 1.0, 15)
    bouquet = Bouquet()

    while True:
        flower_name = input("Enter flower name (or 'done' to finish adding flowers): ")
        if flower_name.lower() == 'done':
            break

        if flower_name.lower() not in ['rose', 'tulip', 'daisy']:
            print("Invalid flower name. Choose from 'Rose', 'Tulip', or 'Daisy'.")
            continue

        quantity = int(input(f"Enter quantity of {flower_name}: "))
        if quantity < 0:
            print("Invalid quantity. Quantity must be a positive integer.")
            continue

        if flower_name.lower() == 'rose':
            bouquet.add_flower(rose, quantity)
        elif flower_name.lower() == 'tulip':
            bouquet.add_flower(tulip, quantity)
        elif flower_name.lower() == 'daisy':
            bouquet.add_flower(daisy, quantity)

    # Sell the bouquet
    if bouquet.sell():
        print("Bouquet sold successfully.")
        print(f"Total Price: ${bouquet.calculate_total_price()}")
    else:
        print("Bouquet sale failed.")
    while True:
        flower_name = input("Enter flower name to order more (or 'done' to finish ordering): ")
        if flower_name.lower() == 'done':
            break

        if flower_name.lower() not in ['rose', 'tulip', 'daisy']:
            print("Invalid flower name. Choose from 'Rose', 'Tulip', or 'Daisy'.")
            continue

        quantity = int(input(f"Enter quantity to order for {flower_name}: "))
        if quantity < 0:
            print("Invalid quantity. Quantity must be a positive integer.")
            continue

        if flower_name.lower() == 'rose':
            rose.order_more(quantity)
        elif flower_name.lower() == 'tulip':
            tulip.order_more(quantity)
        elif flower_name.lower() == 'daisy':
            daisy.order_more(quantity)

    print("\nFlower Inventory:")
    print(rose)
    print(tulip)
    print(daisy)
```
